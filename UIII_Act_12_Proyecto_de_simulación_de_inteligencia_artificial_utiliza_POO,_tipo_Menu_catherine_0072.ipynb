```python
# ==========================================
# IA DE PANADEROS
# PROGRAMACIÓN ORIENTADA A OBJETOS
# ==========================================

# Clase en singular
class Panadero:

    # Constructor
    def __init__(self, tipos_pan, horno, produccion,
                 panaderia, receta, horario):

        self.tipos_pan = tipos_pan
        self.horno = horno
        self.produccion = produccion
        self.panaderia = panaderia
        self.receta = receta
        self.horario = horario

    # ==========================================
    # FUNCION AGREGAR
    # ==========================================
    @staticmethod
    def agregar_panadero(lista_panaderos):

        print("\n🥖 AGREGAR PANADERO 🥖")

        tipos_pan = input("Tipos de pan: ")
        horno = input("Horno utilizado: ")
        produccion = input("Producción diaria: ")
        panaderia = input("Nombre de la panadería: ")
        receta = input("Receta secreta: ")
        horario = input("Horario de madrugada: ")

        nuevo_panadero = Panadero(
            tipos_pan,
            horno,
            produccion,
            panaderia,
            receta,
            horario
        )

        lista_panaderos.append(nuevo_panadero)

        print("\n✅ Panadero agregado correctamente.")

    # ==========================================
    # FUNCION LISTAR
    # ==========================================
    @staticmethod
    def listar_panadero(lista_panaderos):

        print("\n📋 LISTA DE PANADEROS 📋")

        if len(lista_panaderos) == 0:
            print("No hay panaderos registrados.")

        else:

            print("=" * 120)

            print(f"{'No.':<5}"
                  f"{'TIPOS DE PAN':<20}"
                  f"{'HORNO':<20}"
                  f"{'PRODUCCIÓN':<15}"
                  f"{'PANADERÍA':<20}"
                  f"{'RECETA':<20}"
                  f"{'HORARIO':<20}")

            print("=" * 120)

            contador = 1

            for panadero in lista_panaderos:

                print(f"{contador:<5}"
                      f"{panadero.tipos_pan:<20}"
                      f"{panadero.horno:<20}"
                      f"{panadero.produccion:<15}"
                      f"{panadero.panaderia:<20}"
                      f"{panadero.receta:<20}"
                      f"{panadero.horario:<20}")

                contador += 1

            print("=" * 120)

    # ==========================================
    # FUNCION ACTUALIZAR
    # ==========================================
    @staticmethod
    def actualizar_panadero(lista_panaderos):

        print("\n✏️ ACTUALIZAR PANADERO ✏️")

        Panadero.listar_panadero(lista_panaderos)

        if len(lista_panaderos) > 0:

            posicion = int(input("Número del panadero a actualizar: "))

            if posicion >= 1 and posicion <= len(lista_panaderos):

                panadero = lista_panaderos[posicion - 1]

                print("\nIngrese los nuevos datos")

                panadero.tipos_pan = input("Tipos de pan: ")
                panadero.horno = input("Horno utilizado: ")
                panadero.produccion = input("Producción diaria: ")
                panadero.panaderia = input("Nombre de la panadería: ")
                panadero.receta = input("Receta secreta: ")
                panadero.horario = input("Horario de madrugada: ")

                print("\n✅ Panadero actualizado correctamente.")

            else:
                print("❌ Número incorrecto.")

    # ==========================================
    # FUNCION BORRAR
    # ==========================================
    @staticmethod
    def borrar_panadero(lista_panaderos):

        print("\n🗑️ BORRAR PANADERO 🗑️")

        Panadero.listar_panadero(lista_panaderos)

        if len(lista_panaderos) > 0:

            posicion = int(input("Número del panadero a borrar: "))

            if posicion >= 1 and posicion <= len(lista_panaderos):

                lista_panaderos.pop(posicion - 1)

                print("\n✅ Panadero eliminado correctamente.")

            else:
                print("❌ Número incorrecto.")


# ==========================================
# LISTA PRINCIPAL
# ==========================================
lista_panaderos = []


# ==========================================
# MENU PRINCIPAL
# ==========================================
while True:

    print("\n")
    print("===================================")
    print("🤖 IA DE PANADEROS")
    print("===================================")
    print("1. Agregar panadero")
    print("2. Listar panaderos")
    print("3. Actualizar panadero")
    print("4. Borrar panadero")
    print("5. Salir")
    print("===================================")

    opcion = input("Seleccione una opción: ")

    # IF - ELIF - ELSE
    if opcion == "1":
        Panadero.agregar_panadero(lista_panaderos)

    elif opcion == "2":
        Panadero.listar_panadero(lista_panaderos)

    elif opcion == "3":
        Panadero.actualizar_panadero(lista_panaderos)

    elif opcion == "4":
        Panadero.borrar_panadero(lista_panaderos)

    elif opcion == "5":
        print("\n👋 Gracias por usar la IA de Panaderos.")
        break

    else:
        print("\n❌ Opción no válida.")
```
